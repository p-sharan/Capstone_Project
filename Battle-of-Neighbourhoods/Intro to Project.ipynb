# Comparing venue sites for different  American cities

### Introduction

This project aims to compare different venues for different american cities and neighbourhoods within by showing  how many restaurants are there in a particular city and of what type.
This kind of data may help people who wish to move or visit a particular neigbhourhood and people who wish to open a new shop. 

### Data

The data used in this project is mainly from Foursquare API that provides location data for venues and neighbourhoods. The geopy library is also used to get longitude and latitude data for specific locations and merged with the foursquare venue data.
